# Forecast Accuracy: Baseline vs. New Methodology

Two horizontal bar charts (Beverages and Snacks), comparing baseline accuracy (original model) against the final accuracy (new methodology), with a dotted vertical line marking the industry benchmark (75%).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data: (Category, Product, Channel, Baseline %, Final %)
data = [
    ("Beverages", "Natural Juice", "E-commerce", 9.1, 61.2),
    ("Beverages", "Energy Drink", "Traditional", 68.6, 86.0),
    ("Beverages", "Flavored Water", "E-commerce", -20.2, -7.3),
    ("Beverages", "Natural Juice", "Traditional", 69.1, 77.6),
    ("Beverages", "Energy Drink", "Supermarkets", 71.4, 74.6),
    ("Beverages", "Flavored Water", "Traditional", 63.9, 63.9),
    ("Beverages", "Flavored Water", "Supermarkets", 66.4, 63.6),
    ("Beverages", "Natural Juice", "Supermarkets", 74.2, 67.7),
    ("Beverages", "Energy Drink", "E-commerce", 5.6, -22.9),
    ("Snacks", "Cereal Bar", "Supermarkets", 71.8, 79.2),
    ("Snacks", "Cereal Bar", "Traditional", 56.0, 61.0),
    ("Snacks", "Cereal Bar", "E-commerce", 79.1, 84.0),
    ("Snacks", "Whole Grain Crackers", "Supermarkets", 69.8, 73.0),
    ("Snacks", "Whole Grain Crackers", "E-commerce", 77.6, 79.9),
    ("Snacks", "Whole Grain Crackers", "Traditional", 60.9, 43.1),
]

BENCHMARK = 75.0
INDUSTRY_BENCHMARK_LABEL = "Industry benchmark (75%)"

COLOR_BASELINE = "#808080"   # gray
COLOR_FINAL = "#1F3864"      # dark blue
COLOR_BENCHMARK = "#404040"  # dark gray
COLOR_TEXT = "#1A1A1A"       # near-black
```

```python
def plot_category(category, filename=None):
    rows = [r for r in data if r[0] == category]
    rows_sorted = sorted(rows, key=lambda r: r[4], reverse=True)
    labels = [f"{p} - {c}" for _, p, c, _, _ in rows_sorted]
    baseline_vals = [r[3] for r in rows_sorted]
    final_vals = [r[4] for r in rows_sorted]

    n = len(labels)
    y = np.arange(n)
    bar_height = 0.36

    fig, ax = plt.subplots(figsize=(9, 0.75 * n + 2))

    bars_baseline = ax.barh(y + bar_height/2, baseline_vals, height=bar_height,
                             color=COLOR_BASELINE, label="Baseline (original model)")
    bars_final = ax.barh(y - bar_height/2, final_vals, height=bar_height,
                          color=COLOR_FINAL, label="Final (new methodology)")

    # Percentage labels at the end of each bar
    label_pad = 1.2
    for bar, val in zip(bars_baseline, baseline_vals):
        x = bar.get_width()
        ha = "left" if val >= 0 else "right"
        ax.text(x + (label_pad if val >= 0 else -label_pad), bar.get_y() + bar.get_height()/2,
                f"{val:.1f}%", va="center", ha=ha, fontsize=8.5, color=COLOR_TEXT)
    for bar, val in zip(bars_final, final_vals):
        x = bar.get_width()
        ha = "left" if val >= 0 else "right"
        ax.text(x + (label_pad if val >= 0 else -label_pad), bar.get_y() + bar.get_height()/2,
                f"{val:.1f}%", va="center", ha=ha, fontsize=8.5, color=COLOR_TEXT, fontweight="bold")

    ax.axvline(BENCHMARK, color=COLOR_BENCHMARK, linestyle=":", linewidth=1.8, zorder=3)
    # Ancorado em coordenada de eixo (nao de dado) -- fica sempre acima de todas as
    # barras, independente de quantas houver, sem sobrepor nada.
    ax.text(BENCHMARK, 1.01, INDUSTRY_BENCHMARK_LABEL, color=COLOR_BENCHMARK,
             fontsize=9, ha="center", va="bottom", transform=ax.get_xaxis_transform())

    ax.set_yticks(y)
    ax.set_yticklabels(labels, fontsize=10, color=COLOR_TEXT)
    ax.invert_yaxis()

    # Extra x-range so the end-of-bar labels don't get clipped
    all_vals = baseline_vals + final_vals
    ax.set_xlim(min(all_vals) - 10, max(all_vals) + 10)

    ax.set_xlabel("Accuracy (1 - MAPE)", fontsize=11, color=COLOR_TEXT)
    ax.set_title(f"Forecast accuracy - {category}: Baseline vs. New Methodology",
                 fontsize=13, color=COLOR_TEXT, pad=15)

    ax.xaxis.set_major_formatter(lambda v, pos: f"{v:.0f}%")
    ax.tick_params(colors=COLOR_TEXT)
    ax.spines[["top", "right"]].set_visible(False)
    ax.spines[["left", "bottom"]].set_color("#808080")
    ax.grid(axis="x", color="#D9D9D9", linewidth=0.6, zorder=0)
    ax.set_axisbelow(True)

    ax.legend(loc="lower right", fontsize=9, frameon=False, labelcolor=COLOR_TEXT)

    fig.tight_layout()
    if filename:
        fig.savefig(filename, dpi=200, bbox_inches="tight", facecolor="white")
        print(f"Chart saved to {filename}")
    plt.show()
```

## Beverages

```python
plot_category("Beverages", "accuracy_beverages.png")
```

## Snacks

```python
plot_category("Snacks", "accuracy_snacks.png")
```
